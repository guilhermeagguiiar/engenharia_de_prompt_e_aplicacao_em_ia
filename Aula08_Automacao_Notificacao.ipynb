Crie um programa autonomo que dispare alertas (mensagens de alerta no console ou e-mail simulado) unicamente quando uma condição critica for atingida

```python
import time
import random

# --- Configuração ---
# Exemplo: valor acima de 90 é considerado crítico
CRITICAL_THRESHOLD = 90
# Intervalo de tempo entre cada verificação (em segundos)
CHECK_INTERVAL_SECONDS = 5
# Destinatário do e-mail simulado
EMAIL_RECIPIENT = "admin@empresa.com"

# --- Funções de Alerta ---
def send_console_alert(message):
    """Envia uma mensagem de alerta para o console."""
    print(f"[ALERTA CONSOLE] {message}")

def send_simulated_email_alert(subject, body, recipient):
    """Simula o envio de um alerta por e-mail."""
    print(f"[ALERTA E-MAIL SIMULADO] Para: {recipient}, Assunto: {subject}, Corpo: {body}")

# --- Programa Principal ---
print("Iniciando programa de monitoramento autônomo...")
print(f"Monitorando condição crítica (valor > {CRITICAL_THRESHOLD}) a cada {CHECK_INTERVAL_SECONDS} segundos.")
print("Pressione Ctrl+C para encerrar a qualquer momento.\n")

critical_condition_met_previously = False # Flag para rastrear se a condição crítica foi atingida na verificação anterior

try:
    while True:
        # Simula uma métrica ou condição em tempo real.
        # Em um cenário real, isso viria de sensores, APIs, bancos de dados, etc.
        current_value = random.randint(0, 100) # Simula um valor entre 0 e 100
        print(f"[{time.strftime('%H:%M:%S')}] Valor atual monitorado: {current_value}")

        if current_value > CRITICAL_THRESHOLD:
            if not critical_condition_met_previously:
                # A condição se tornou crítica AGORA
                message = f"Condição CRÍTICA atingida! Valor atual: {current_value} (limite: {CRITICAL_THRESHOLD})"
                send_console_alert(message)
                send_simulated_email_alert("ALERTA CRÍTICO: Condição Excedida", message, EMAIL_RECIPIENT)
                critical_condition_met_previously = True
            else:
                # A condição continua crítica, mas o alerta já foi disparado
                print("Condição crítica persiste, alerta já disparado.")
        else:
            if critical_condition_met_previously:
                # A condição retornou ao normal APÓS ter sido crítica
                message = f"Condição retornou ao normal. Valor atual: {current_value}"
                send_console_alert(message)
                send_simulated_email_alert("INFORMAÇÃO: Condição Normalizada", message, EMAIL_RECIPIENT)
                critical_condition_met_previously = False
            else:
                # A condição está normal e não era crítica antes
                print("Condição normal. Nenhum alerta necessário.")

        time.sleep(CHECK_INTERVAL_SECONDS)

except KeyboardInterrupt:
    print("\nPrograma de monitoramento encerrado pelo usuário.")
except Exception as e:
    print(f"\nOcorreu um erro inesperado: {e}")
```
