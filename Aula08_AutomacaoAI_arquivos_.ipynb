```python
import os           # Biblioteca para interagir com o sistema operacional (caminhos, diretórios)
import shutil       # Biblioteca para operações de alto nível em arquivos e diretórios (mover, copiar, deletar)

# --- 1. Configurações Iniciais ---
# Define o diretório onde os arquivos bagunçados serão criados. Usamos um nome diferente para evitar conflito
# com execuções anteriores se o diretório 'bagunca' ainda existir parcialmente.
BAGUNCA_DIR = 'bagunca_refatorado'

# Define as extensões de arquivo que queremos testar, conforme especificado no pedido.
extensions = ['.pdf', '.jpg', '.xls', '.doc', '.py', '.ipynb', '.txt', '.png', '.csv', '.zip']

# Número de arquivos a serem criados para cada tipo de extensão, conforme especificado.
num_files_per_extension = 10

# --- 2. Criação e Limpeza do Diretório de Teste ---
print(f"\n--- Preparando o diretório de trabalho: '{BAGUNCA_DIR}' ---")

# Verifica se o diretório principal para os arquivos 'bagunçados' já existe.
if not os.path.exists(BAGUNCA_DIR):
    # Se não existir, cria o diretório.
    os.makedirs(BAGUNCA_DIR)
    print(f"Diretório '{BAGUNCA_DIR}' criado com sucesso para a organização.")
else:
    # Se já existir, informa que ele será utilizado.
    print(f"Diretório '{BAGUNCA_DIR}' já existe. Será utilizado para a operação.")

# Limpa qualquer conteúdo existente dentro do diretório 'bagunca_refatorado'
# para garantir um ambiente de teste limpo a cada execução.
print(f"Limpando o conteúdo existente de '{BAGUNCA_DIR}' antes de gerar novos arquivos...")
for item in os.listdir(BAGUNCA_DIR): # Lista todos os arquivos e diretórios dentro de BAGUNCA_DIR.
    item_path = os.path.join(BAGUNCA_DIR, item) # Constrói o caminho completo para cada item.
    try:
        if os.path.isfile(item_path) or os.path.islink(item_path): # Se for um arquivo ou link simbólico...
            os.unlink(item_path) # ...remove o arquivo/link.
        elif os.path.isdir(item_path): # Se for um diretório...
            shutil.rmtree(item_path)   # ...remove o diretório e todo o seu conteúdo (recursivamente).
    except Exception as e:
        # Captura e exibe qualquer erro que possa ocorrer durante a exclusão.
        print(f'Falha ao deletar {item_path}. Razão: {e}')
print(f"Conteúdo existente de '{BAGUNCA_DIR}' limpo com sucesso.\n")

# --- 3. Geração de Arquivos de Exemplo ---
print("--- Gerando arquivos de exemplo para simular a 'bagunça' ---")
# Itera sobre cada extensão de arquivo definida.
for ext in extensions:
    # Para cada extensão, cria o número especificado de arquivos.
    for i in range(num_files_per_extension):
        # Constrói o nome do arquivo, ex: 'arquivo_pdf_0.pdf'
        filename = f"arquivo_{ext[1:]}_{i}{ext}"
        # Constrói o caminho completo onde o arquivo será criado dentro de BAGUNCA_DIR.
        filepath = os.path.join(BAGUNCA_DIR, filename)
        # Abre o arquivo no modo de escrita ('w') e adiciona um conteúdo simples.
        with open(filepath, 'w') as f:
            f.write(f"Conteúdo de {filename}\nEste é um arquivo de teste com extensão {ext}.")
        # O print abaixo está comentado para não poluir a saída, mas pode ser descomentado para depuração.
        # print(f"Criado: {filepath}")
print(f"Total de {len(extensions) * num_files_per_extension} arquivos de exemplo criados em '{BAGUNCA_DIR}'.\n")

# --- 4. Organização dos Arquivos ---
print(f"--- Iniciando a organização automática dos arquivos em '{BAGUNCA_DIR}' ---")
# Itera sobre todos os arquivos e diretórios presentes no BAGUNCA_DIR após a geração.
for filename in os.listdir(BAGUNCA_DIR):
    # Constrói o caminho completo do arquivo atual.
    file_path = os.path.join(BAGUNCA_DIR, filename)

    # Verifica se o item atual é de fato um arquivo (ignora diretórios e links simbólicos por enquanto).
    if os.path.isfile(file_path):
        # Separa o nome do arquivo da sua extensão. Ex: 'documento.pdf' -> ('documento', '.pdf')
        _, file_extension = os.path.splitext(filename)
        # Converte a extensão para minúsculas para garantir consistência (ex: '.PDF' e '.pdf' vão para a mesma pasta).
        file_extension = file_extension.lower()

        # Se o arquivo não tiver uma extensão (ex: 'Makefile'), ele será ignorado por este programa.
        if not file_extension:
            print(f"Aviso: Arquivo '{filename}' não possui extensão conhecida, será ignorado.")
            continue # Pula para o próximo arquivo no loop.

        # Remove o ponto inicial da extensão para usá-la como nome da pasta de destino. Ex: '.pdf' -> 'pdf'
        target_folder_name = file_extension[1:]
        # Constrói o caminho completo para a nova pasta de destino. Ex: 'bagunca_refatorado/pdf'
        target_folder_path = os.path.join(BAGUNCA_DIR, target_folder_name)

        # Verifica se a pasta de destino para esta extensão já existe.
        if not os.path.exists(target_folder_path):
            # Se não existir, cria a nova pasta.
            os.makedirs(target_folder_path)
            print(f"Pasta de destino criada: '{target_folder_name}/'")

        # Constrói o caminho completo para onde o arquivo será movido (nova localização).
        destination_path = os.path.join(target_folder_path, filename)

        # Move o arquivo do caminho original para o caminho de destino.
        shutil.move(file_path, destination_path)
        print(f"Movido: '{filename}' para '{target_folder_name}/'")

print("Organização de arquivos concluída com sucesso!\n")

# --- 5. Verificação Pós-Organização (Opcional) ---
print(f"--- Estrutura de diretórios e arquivos em '{BAGUNCA_DIR}' após a organização ---")
# A função os.walk() percorre a árvore de diretórios, gerando os nomes dos diretórios e arquivos.
for root, dirs, files in os.walk(BAGUNCA_DIR):
    # Calcula o nível de indentação para representar a profundidade na árvore de diretórios.
    # Isso ajuda a visualizar a hierarquia de forma clara.
    level = root.replace(BAGUNCA_DIR, '').count(os.sep)
    indent = ' ' * 4 * (level)
    # Imprime o nome do diretório atual, com a indentação apropriada.
    print(f'{indent}{os.path.basename(root)}/')
    # Calcula a indentação para os arquivos dentro do diretório atual.
    subindent = ' ' * 4 * (level + 1)
    # Itera e imprime cada arquivo encontrado no diretório atual, com a indentação dos subitens.
    for f in files:
        print(f'{subindent}{f}')
print("\n--- Fim da execução do programa refatorado ---")

# --- 6. Limpeza Final (Totalmente Opcional - Descomente para ativar) ---
# Esta seção está comentada por segurança. Se você deseja remover o diretório 'bagunca_refatorado'
# e todo o seu conteúdo após a execução, descomente as linhas abaixo. CUIDADO: esta ação é irreversível!

# print(f"\n--- Tentando limpar o diretório '{BAGUNCA_DIR}' (requer descomentar o código) ---")
# if os.path.exists(BAGUNCA_DIR): # Verifica se o diretório existe antes de tentar removê-lo.
#     shutil.rmtree(BAGUNCA_DIR)   # Remove o diretório e todos os seus subdiretórios e arquivos.
#     print(f"Diretório '{BAGUNCA_DIR}' e seu conteúdo removidos com sucesso.")
# else:
#     print(f"Diretório '{BAGUNCA_DIR}' não encontrado para remoção.")
```
