## Organização Automática de Arquivos por Extensão

Este programa demonstra como organizar arquivos em pastas baseadas em suas extensões, utilizando as bibliotecas `os` e `shutil`.

### 1. Importando as Bibliotecas Necessárias

Importamos `os` para interagir com o sistema operacional (criar diretórios, listar arquivos) e `shutil` para operações de alto nível com arquivos (mover arquivos).

```python
import os           # Biblioteca para interagir com o sistema operacional (caminhos, diretórios)
import shutil       # Biblioteca para operações de alto nível em arquivos e diretórios (mover, copiar, deletar)
import random       # Biblioteca para gerar números aleatórios (não essencial para a organização, mas útil para simulação)
```

### 2. Configurações Iniciais e Criação de Arquivos de Exemplo

Vamos definir algumas extensões e o número de arquivos para cada uma. Em seguida, criaremos uma pasta 'bagunca' e geraremos arquivos fictícios para simular um diretório desorganizado. Serão criados 10 arquivos de cada tipo.

```python
# Define o diretório onde os arquivos bagunçados serão criados
BAGUNCA_DIR = 'bagunca'

# Define as extensões de arquivo que queremos testar
extensions = ['.pdf', '.jpg', '.xls', '.doc', '.py', '.ipynb', '.txt', '.png', '.csv', '.zip']

# Número de arquivos por tipo de extensão, conforme solicitado (10 arquivos de cada tipo)
num_files_per_extension = 10

# Cria o diretório 'bagunca' se ele não existir
if not os.path.exists(BAGUNCA_DIR): # Verifica se o diretório já existe
    os.makedirs(BAGUNCA_DIR)       # Cria o diretório se não existir
    print(f"Diretório '{BAGUNCA_DIR}' criado com sucesso.")
else:
    print(f"Diretório '{BAGUNCA_DIR}' já existe.")

# Limpa o diretório 'bagunca' antes de criar novos arquivos para evitar duplicatas em execuções repetidas
print(f"Limpando o conteúdo existente de '{BAGUNCA_DIR}'...")
for filename in os.listdir(BAGUNCA_DIR): # Lista todos os itens dentro do diretório
    file_path = os.path.join(BAGUNCA_DIR, filename) # Constrói o caminho completo para o item
    try:
        if os.path.isfile(file_path) or os.path.islink(file_path): # Se for um arquivo ou link simbólico
            os.unlink(file_path) # Remove o arquivo ou link
        elif os.path.isdir(file_path): # Se for um diretório
            shutil.rmtree(file_path)   # Remove o diretório e todo o seu conteúdo recursivamente
    except Exception as e:
        print(f'Falha ao deletar {file_path}. Razão: {e}') # Captura e exibe erros durante a exclusão
print(f"Conteúdo existente de '{BAGUNCA_DIR}' limpo.")

# Gera arquivos fictícios para cada extensão
print("Gerando arquivos de exemplo...")
for ext in extensions: # Itera sobre cada extensão definida
    for i in range(num_files_per_extension): # Cria 'num_files_per_extension' arquivos para cada extensão
        filename = f"arquivo_{ext[1:]}_{i}{ext}" # Ex: 'arquivo_pdf_0.pdf'
        filepath = os.path.join(BAGUNCA_DIR, filename) # Constrói o caminho completo para o novo arquivo
        with open(filepath, 'w') as f: # Abre o arquivo em modo de escrita ('w')
            f.write(f"Conteúdo de {filename}") # Escreve um conteúdo simples no arquivo
        # print(f"Criado: {filepath}") # Comentado para evitar muita saída no console
print(f"Total de {len(extensions) * num_files_per_extension} arquivos de exemplo criados em '{BAGUNCA_DIR}'.")
```

### 3. Programa Principal: Organizando os Arquivos

Este é o coração do programa. Ele irá:
1. Listar todos os arquivos no diretório 'bagunca'.
2. Para cada arquivo, extrair sua extensão.
3. Criar uma pasta com o nome da extensão (se ainda não existir).
4. Mover o arquivo para essa nova pasta.

```python
print(f"Iniciando a organização dos arquivos em '{BAGUNCA_DIR}'...")

# Itera sobre todos os arquivos e diretórios no BAGUNCA_DIR
for filename in os.listdir(BAGUNCA_DIR): # Lista todos os arquivos e pastas no diretório bagunça
    # Constrói o caminho completo para o arquivo
    file_path = os.path.join(BAGUNCA_DIR, filename) # Junta o nome do diretório base com o nome do arquivo

    # Verifica se é um arquivo (e não um diretório ou link simbólico)
    if os.path.isfile(file_path): # Garante que estamos lidando apenas com arquivos
        # Obtém o nome base e a extensão do arquivo
        # Ex: 'documento.pdf' -> ('documento', '.pdf')
        _, file_extension = os.path.splitext(filename) # Separa o nome do arquivo da extensão

        # Garante que a extensão esteja em minúsculas para padronização
        file_extension = file_extension.lower() # Converte a extensão para minúsculas para evitar pastas duplicadas (e.g., '.Pdf' e '.pdf')

        # Se não houver extensão (ex: 'meuarquivo'), pulamos ou podemos criar uma pasta 'sem_extensao'
        if not file_extension: # Verifica se o arquivo tem uma extensão
            print(f"Arquivo '{filename}' não possui extensão, pulando.") # Informa que o arquivo será ignorado
            continue # Pula para o próximo arquivo

        # Remove o ponto inicial da extensão para usar como nome da pasta
        # Ex: '.pdf' -> 'pdf'
        target_folder_name = file_extension[1:] # Remove o ponto inicial da extensão para usar como nome da pasta (ex: 'pdf', 'jpg')

        # Constrói o caminho completo para a nova pasta de destino
        target_folder_path = os.path.join(BAGUNCA_DIR, target_folder_name) # Cria o caminho completo para a pasta de destino (e.g., 'bagunca/pdf')

        # Cria a pasta de destino se ela não existir
        if not os.path.exists(target_folder_path): # Verifica se a pasta de destino já existe
            os.makedirs(target_folder_path)       # Cria a pasta se ela não existir
            print(f"Pasta '{target_folder_name}' criada.") # Informa que uma nova pasta foi criada

        # Constrói o caminho completo de destino do arquivo
        destination_path = os.path.join(target_folder_path, filename) # Caminho final onde o arquivo será movido

        # Move o arquivo para a pasta de destino
        shutil.move(file_path, destination_path) # Move o arquivo do caminho original para o caminho de destino
        print(f"Movido: '{filename}' para '{target_folder_name}/'") # Confirma o movimento do arquivo

print("Organização concluída!")
```

### 4. Verificação (Opcional)

Vamos listar o conteúdo do diretório 'bagunca' para ver as pastas criadas e alguns de seus conteúdos.

```python
print(f"Conteúdo de '{BAGUNCA_DIR}' após a organização:")
# os.walk() gera os nomes dos arquivos em uma árvore de diretórios
for root, dirs, files in os.walk(BAGUNCA_DIR):
    # Calcula o nível de indentação para uma visualização hierárquica
    level = root.replace(BAGUNCA_DIR, '').count(os.sep)
    indent = ' ' * 4 * (level)
    # Imprime o nome do diretório atual, com indentação
    print(f'{indent}{os.path.basename(root)}/')
    # Calcula a indentação para os arquivos dentro do diretório
    subindent = ' ' * 4 * (level + 1)
    # Itera e imprime cada arquivo encontrado no diretório atual
    for f in files:
        print(f'{subindent}{f}')
```

### 6. Programa Refatorado em Um Único Bloco de Código

Este bloco de código integra todas as etapas anteriores (importações, configuração, criação de arquivos de exemplo, organização e verificação) em uma única execução para maior concisão.

```python
import os           # Biblioteca para interagir com o sistema operacional (caminhos, diretórios)
import shutil       # Biblioteca para operações de alto nível em arquivos e diretórios (mover, copiar, deletar)

# --- 1. Configurações Iniciais ---
BAGUNCA_DIR = 'bagunca_refatorado'
extensions = ['.pdf', '.jpg', '.xls', '.doc', '.py', '.ipynb', '.txt', '.png', '.csv', '.zip']
num_files_per_extension = 10

# --- 2. Criação e Limpeza do Diretório de Teste ---
print(f"\n--- Preparando o diretório '{BAGUNCA_DIR}' ---")
# Cria o diretório 'bagunca_refatorado' se ele não existir
if not os.path.exists(BAGUNCA_DIR):
    os.makedirs(BAGUNCA_DIR)
    print(f"Diretório '{BAGUNCA_DIR}' criado com sucesso.")
else:
    print(f"Diretório '{BAGUNCA_DIR}' já existe.")

# Limpa o conteúdo existente do diretório 'bagunca_refatorado'
print(f"Limpando o conteúdo existente de '{BAGUNCA_DIR}'...")
for filename in os.listdir(BAGUNCA_DIR):
    file_path = os.path.join(BAGUNCA_DIR, filename)
    try:
        if os.path.isfile(file_path) or os.path.islink(file_path):
            os.unlink(file_path)
        elif os.path.isdir(file_path):
            shutil.rmtree(file_path)
    except Exception as e:
        print(f'Falha ao deletar {file_path}. Razão: {e}')
print(f"Conteúdo existente de '{BAGUNCA_DIR}' limpo.\n")

# --- 3. Geração de Arquivos de Exemplo ---
print("--- Gerando arquivos de exemplo ---")
for ext in extensions:
    for i in range(num_files_per_extension):
        filename = f"arquivo_{ext[1:]}_{i}{ext}"
        filepath = os.path.join(BAGUNCA_DIR, filename)
        with open(filepath, 'w') as f:
            f.write(f"Conteúdo de {filename}")
print(f"Total de {len(extensions) * num_files_per_extension} arquivos de exemplo criados em '{BAGUNCA_DIR}'.\n")

# --- 4. Organização dos Arquivos ---
print(f"--- Iniciando a organização dos arquivos em '{BAGUNCA_DIR}' ---")
for filename in os.listdir(BAGUNCA_DIR):
    file_path = os.path.join(BAGUNCA_DIR, filename)

    if os.path.isfile(file_path):
        _, file_extension = os.path.splitext(filename)
        file_extension = file_extension.lower()

        if not file_extension:
            print(f"Arquivo '{filename}' não possui extensão, pulando.")
            continue

        target_folder_name = file_extension[1:]
        target_folder_path = os.path.join(BAGUNCA_DIR, target_folder_name)

        if not os.path.exists(target_folder_path):
            os.makedirs(target_folder_path)
            print(f"Pasta '{target_folder_name}' criada.")

        destination_path = os.path.join(target_folder_path, filename)
        shutil.move(file_path, destination_path)
        # print(f"Movido: '{filename}' para '{target_folder_name}/'") # Descomente para ver cada movimento
print("Organização concluída!\n")

# --- 5. Verificação Pós-Organização ---
print(f"--- Conteúdo de '{BAGUNCA_DIR}' após a organização ---")
for root, dirs, files in os.walk(BAGUNCA_DIR):
    level = root.replace(BAGUNCA_DIR, '').count(os.sep)
    indent = ' ' * 4 * (level)
    print(f'{indent}{os.path.basename(root)}/')
    subindent = ' ' * 4 * (level + 1)
    for f in files:
        print(f'{subindent}{f}')
print("\n--- Fim da execução do programa refatorado ---")

# --- 6. Limpeza Final (Opcional - Descomente para ativar) ---
# print(f"\n--- Limpando '{BAGUNCA_DIR}' (descomente o código para ativar) ---")
# if os.path.exists(BAGUNCA_DIR):
#     shutil.rmtree(BAGUNCA_DIR)
#     print(f"Diretório '{BAGUNCA_DIR}' e seu conteúdo removidos.")
# else:
#     print(f"Diretório '{BAGUNCA_DIR}' não encontrado para remoção.")
```
