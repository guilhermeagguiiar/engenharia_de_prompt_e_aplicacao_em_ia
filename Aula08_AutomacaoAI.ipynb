```python
import random

def check_temperature():
    # Solicita ao usuário que insira a temperatura atual do sistema.
    while True:
        try:
            temperature_input = input("Por favor, insira a temperatura atual do sistema em °C: ")
            # Converte a entrada do usuário para um número de ponto flutuante.
            temperature = float(temperature_input)
            break # Sai do loop se a conversão for bem-sucedida
        except ValueError:
            # Trata o erro se a entrada não for um número válido.
            print("Entrada inválida. Por favor, insira um número para a temperatura.")

    # Exibe a temperatura atual formatada com duas casas decimais, usando concatenação de strings.
    print("Temperatura atual do sistema: " + "{:.2f}".format(temperature) + "°C")

    # Verifica se a temperatura ultrapassa o limite de 30°C.
    if temperature > 30:
        # Se a temperatura for maior que 30°C, exibe um alerta de superaquecimento.
        print("ALERTA: Superaquecimento detectado! Temperatura acima de 30°C.")
    else:
        # Caso contrário, informa que a temperatura do sistema está normal.
        print("Temperatura do sistema está normal.")

# Chama a função para executar a verificação da temperatura.
check_temperature()
```
